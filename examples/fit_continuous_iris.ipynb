```python
import phitter
```

```python
df = pandas.read_csv("https://gist.githubusercontent.com/phitterio/6e1cdb92a5f518c7459484c619185fc5/raw/4072c5233d03204d8999943398b23c2393bcd637/iris.csv")
df.head()
```

```python
data = df["SepalWidthCm"].to_numpy()
```

```python
phi = phitter.PHITTER(data=data)
phi.fit(n_workers=2)
```

```python
phi.best_distribution
```

```python
phi.df_sorted_distributions_sse
```
