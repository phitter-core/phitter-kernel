```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy
import scipy.stats
```

```python
mpl.style.use("ggplot")
```

```python
sample_distribution_file = open("../../datasets_test/continuous/data_1000/data_beta.txt", "r")
data = [float(x.replace(",", ".")) for x in sample_distribution_file.read().splitlines()]
sample_distribution_file.close()
```

```python
## KDE
distribution = scipy.stats.gaussian_kde(data)
```

```python
print(distribution.pdf(12.5))
print(distribution.evaluate(12.5))
```

```python
plt.figure(figsize=(8, 4))
plt.hist(data, density=True, ec="white")
plt.title("HISTOGRAM")
plt.xlabel("Values")
plt.ylabel("Frequencies")

x_plot = numpy.linspace(min(data), max(data), 1000)
y_plot = distribution.pdf(x_plot)
plt.plot(x_plot, y_plot, label="PDF KDE")

plt.legend(title="DISTRIBUTIONS", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```
