```python
import numpy
import plotly.graph_objects as go
```

```python
def get_data(path: str) -> list[float | int]:
    sample_distribution_file = open(path, "r")
    data = [float(x.replace(",", ".")) for x in sample_distribution_file.read().splitlines()]
    sample_distribution_file.close()
    return data
```

```python
path = "../../../datasets_test/continuous/data_1000/data_beta.txt"
data = get_data(path)
```

```python
def ecdf(data):
    x = numpy.unique(numpy.sort(data))
    y = numpy.arange(1, len(x) + 1) / len(x)
    return x, y
```

```python
# data = [6.23, 5.58, 7.06, 6.42, 5.20]
```

```python
x, y = ecdf(data)
```

```python
new_x = numpy.repeat(x, 2)[1:-1]
new_y = numpy.repeat(y, 2)[:-1]
```

```python
# fig = px.ecdf(x=data)
# fig.data[0].name = "ECDF"
fig = go.Figure()
fig.add_trace(go.Scatter(x=new_x, y=new_y, mode="lines", name="Empirical Distribution", line=dict(color="firebrick", width=5), showlegend=True))
# fig.add_trace(go.Scatter(x=data, y=data, mode="lines", name="xxx", line=dict(color="firebrick")))
fig.update_layout(
    height=400,
    width=700,
    title="HISTOGRAM",
    xaxis_title="Values",
    yaxis_title="Frequencies",
    # legend_title="LEGEND",
    template="ggplot2",
    xaxis_range=[min(data) - 0.3, max(data) + 0.3],
)
fig.show()
```
