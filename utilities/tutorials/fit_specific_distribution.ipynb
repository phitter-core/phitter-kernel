```python
import phitter
```

```python
distribution = phitter.continuous.BETA(parameters={"alpha": 5, "beta": 2, "A": 100, "B": 1000})
data = distribution.sample(10000)
```

```python
phitter_cont = phitter.PHITTER(data=data, distributions_to_fit=["beta"])
phitter_cont.fit(n_workers=4)
```

```python
phitter_cont.df_sorted_distributions_sse
```

```python
phitter_cont.plot_histogram_distributions()
```
