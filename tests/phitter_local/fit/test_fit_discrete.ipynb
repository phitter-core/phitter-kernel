```python
import sys
sys.path.append("../../../")
import phitter
```

```python
def get_data(path: str) -> list[float | int]:
    sample_distribution_file = open(path, "r", encoding="utf-8-sig")
    data = [float(x.replace(",", ".")) for x in sample_distribution_file.read().splitlines()]
    sample_distribution_file.close()
    return data
```

```python
path = "../../../datasets_test/discrete/book2.txt"
data = get_data(path)
```

```python
phitter_disc = phitter.PHITTER(data, fit_type="discrete")
phitter_disc.fit(n_workers=2)
```
