```python
import sys
sys.path.append("../../")
import phitter
```

```python
def get_data(path: str) -> list[float | int]:
    sample_distribution_file = open(path, "r")
    data = [float(x.replace(",", ".")) for x in sample_distribution_file.read().splitlines()]
    sample_distribution_file.close()
    return data
```

```python
path = "../../datasets_test/continuous/data.txt"
data = get_data(path)
```

```python
phitter_cont = phitter.PHITTER(
    data=data,
    fit_type="continuous",
    num_bins=15,
    confidence_level=0.97,
    minimum_sse=1e-2,
    distributions_to_fit=["beta", "normal", "fatigue_life", "triangular"],
)
```

```python
phitter = phitter.PHITTER(data)
phitter.fit(n_jobs=6)
```
