```python
import sys
sys.path.append("../../")
import phitter
```

```python
def get_data(path: str) -> list[float | int]:
    sample_distribution_file = open(path, "r")
    data = [float(x.replace(",", ".")) for x in sample_distribution_file.read().splitlines()]
    sample_distribution_file.close()
    return data
```

```python
path = "../../datasets_test/discrete/sample_geometric.txt"
data = get_data(path)
```

```python
phitter = phitter.PHITTER(data, fit_type="discrete")
phitter.fit(n_workers=1)
```

```python
phitter.plot_histogram(plot_bar_color="gray")
```

```python
phitter.plot_histogram_distributions(50, 1, plot_bar_color="gray")
```

```python
phitter.plot_distribution("geometric", plot_bar_color="gray", plot_line_color="red")
```

```python
phitter.plot_ecdf(plot_bar_color="gray")
```

```python
phitter.plot_ecdf_distribution("geometric", plot_empirical_bar_color="gray", plot_distribution_line_color="red", plot_distribution_line_width=2)
```

```python
phitter.qq_plot("geometric")
```

```python
phitter.qq_plot_regression("geometric")
```
