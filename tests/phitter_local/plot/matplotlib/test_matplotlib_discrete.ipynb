```python
import sys
sys.path.append("../../../../")
import phitter
```

```python
data = phitter.discrete.BINOMIAL(init_parameters_examples=True).sample(10000)
```

```python
phitter_disc = phitter.PHITTER(data, fit_type="discrete")
phitter_disc.fit(n_workers=1)
```

```python
phitter_disc.plot_histogram(plot_engine="matplotlib")
```

```python
phitter_disc.plot_histogram_distributions(plot_engine="matplotlib")
```

```python
phitter_disc.plot_distribution("hypergeometric", plot_engine="matplotlib")
```

```python
phitter_disc.plot_ecdf(plot_engine="matplotlib")
```

```python
phitter_disc.plot_ecdf_distribution("hypergeometric", plot_engine="matplotlib")
```

```python
phitter_disc.qq_plot("hypergeometric", plot_engine="matplotlib")
```

```python
phitter_disc.qq_plot_regression("hypergeometric", plot_engine="matplotlib")
```
