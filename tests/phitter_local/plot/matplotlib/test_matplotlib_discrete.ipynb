```python
import sys
sys.path.append("../../../../")
import phitter
```

```python
data = phitter.discrete.Binomial(init_parameters_examples=True).sample(10000)
```

```python
phi = phitter.Phitter(data=data, fit_type="discrete")
phi.fit(n_workers=1)
```

```python
phi.plot_histogram(plot_engine="matplotlib")
```

```python
phi.plot_histogram_distributions(plot_engine="matplotlib")
```

```python
phi.plot_distribution("hypergeometric", plot_engine="matplotlib")
```

```python
phi.plot_ecdf(plot_engine="matplotlib")
```

```python
phi.plot_ecdf_distribution("hypergeometric", plot_engine="matplotlib")
```

```python
phi.qq_plot("hypergeometric", plot_engine="matplotlib")
```

```python
phi.qq_plot_regression("hypergeometric", plot_engine="matplotlib")
```
